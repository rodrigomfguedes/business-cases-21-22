## <font color='#5D6D7E '>General Context</font> <a class="anchor" id="first-bullet"></a>
  [Back to TOC](#toc)
  
Forecasting is a technique used to predict future events based on historical data. The quality of these predictions can then be compared to what actually happens. For example, sports and politics betting is a form of forecasting. In this scenario, bettors are paid based on the outcome of their predictions. In its statistical sense, forecasting approaches typically use time series,
crosssectional, or longitudinal data.

## <font color='#5D6D7E'>Import and Transform the format of Data</font> <a class="anchor" id="fourth-bullet"></a>
  [Back to TOC](#toc)

#### Data Fields:

For each Dataset of each cryptocurrencies the following features present are:
__`ADR`__: Lowest price during a day <br>
High: Highest price during a day
Open: Price at the start of the day
Close: Price at the end of the day
Adj. Close: Closing price after adjustments for all applicable splits and
dividend distributions. Data is adjusted using
appropriate split and
dividend multipliers, adhering to Center for Research in Security Prices
(CRSP) standards.
Volume: Amount of an asset or security that changes hands over the course of
a day


```python
df_adjClose = pd.read_csv('../data/adj_close.csv')
df_close = pd.read_csv('../data/close.csv')
df_high = pd.read_csv('../data/high.csv')
df_low = pd.read_csv('../data/low.csv')
df_open = pd.read_csv('../data/open.csv')
df_volume = pd.read_csv('../data/volume.csv')
```

```python
#DF LIST TO MERGE
dfs = [df_adjClose, df_close, df_high, df_low, df_open, df_volume]

#GET THE LIST OF THE CRYPTO CURRENCIES
crypto_list = df_adjClose.columns[1:]

#LIST OF MARKET METRICS/OPERATIONS
market_metrics = ['ADJCLOSE','CLOSE','HIGH','LOW','OPEN','VOLUME']
```

```python
#RENAME THE COLUMNS IN THE DF'S WITH THE RESPECTIVE METRIC
i = 0
for df in dfs:
    for crypto in crypto_list:
        df.rename(columns = {crypto : crypto + '_' + market_metrics[i]}, inplace = True)
    i += 1
```

```python
#DATAFRAME WITH ALL OF THE MERGED INFORMATION
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='inner'), dfs)
merged_columns = df_merged.columns
```

```python
df_merged
```

```python
#GET THE INDIVIDUAL DATAFRAMES FOR EVERY CRYPTOCURRENCY
crypto_df_list = []

for crypto in crypto_list:
    
    #taking input as a string
    str = "df_" + crypto[:-4]
    
    #Add the df name to the list
    crypto_df_list.append(str)
    
    #Get the columns for the corresponding crypto
    crypto_columns = ['Date']
    for column in merged_columns:
        if crypto[:-4] in column:
            crypto_columns.append(column)
            
    #Assign the df
    locals()[str] = df_merged[crypto_columns]
```

```python
##EXTRACT THE DF'S INTO CSV'S FOR PRE-PROCESSING/ANALYSIS IN SINGULAR NOTEBOOKS

for df in crypto_df_list:
    locals()[df].to_csv('../data/data_aux/' + df +'.csv',index=False)
```
