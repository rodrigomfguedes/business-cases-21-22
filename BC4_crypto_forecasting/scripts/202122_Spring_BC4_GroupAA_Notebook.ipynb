```python
df_adjClose = pd.read_csv('../data/adj_close.csv')
df_close = pd.read_csv('../data/close.csv')
df_high = pd.read_csv('../data/high.csv')
df_low = pd.read_csv('../data/low.csv')
df_open = pd.read_csv('../data/open.csv')
df_volume = pd.read_csv('../data/volume.csv')
```

```python
#DF LIST TO MERGE
dfs = [df_adjClose, df_close, df_high, df_low, df_open, df_volume]

#GET THE LIST OF THE CRYPTO CURRENCIES
crypto_list = df_adjClose.columns[1:]

#LIST OF MARKET METRICS/OPERATIONS
market_metrics = ['ADJCLOSE','CLOSE','HIGH','LOW','OPEN','VOLUME']
```

```python
#RENAME THE COLUMNS IN THE DF'S WITH THE RESPECTIVE METRIC
i = 0
for df in dfs:
    for crypto in crypto_list:
        df.rename(columns = {crypto : crypto + '_' + market_metrics[i]}, inplace = True)
    i += 1
```

```python
#DATAFRAME WITH ALL OF THE MERGED INFORMATION
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='inner'), dfs)
merged_columns = df_merged.columns
```

```python
df_merged
```

```python
#GET THE INDIVIDUAL DATAFRAMES FOR EVERY CRYPTOCURRENCY
crypto_df_list = []

for crypto in crypto_list:
    
    #taking input as a string
    str = "df_" + crypto[:-4]
    
    #Add the df name to the list
    crypto_df_list.append(str)
    
    #Get the columns for the corresponding crypto
    crypto_columns = ['Date']
    for column in merged_columns:
        if crypto[:-4] in column:
            crypto_columns.append(column)
            
    #Assign the df
    locals()[str] = df_merged[crypto_columns]
```

```python
##EXTRACT THE DF'S INTO CSV'S FOR PRE-PROCESSING/ANALYSIS IN SINGULAR NOTEBOOKS

for df in crypto_df_list:
    locals()[df].to_csv('../data/data_aux/' + df +'.csv',index=False)
```

#### __`Seasonal Reservations`__

```python
#Lists that defined seasons based on month group

summer = ['June','July','August']
spring = ['March','April','May'] 
winter = ['December','January','February']
autumn = ['September','October','November']

seasonsdict={'Summer': summer, 'Spring': spring, 'Winter': winter, 'Autumn':autumn}

#Function to get season based on month
def get_season(month, seasonsdict):
    
    for season_name, season_months in seasonsdict.items():
        if month in season_months:
            return season_name
```

```python
#Create a new column where you store the season of the reservation
df['ArrivalDateSeason'] = df['ArrivalDateMonth'].apply(get_season, args=[seasonsdict])

#Bar chart of seasonal reservations by percentage
(df['ArrivalDateSeason'].value_counts(normalize=True)*100).plot.bar(color = '#7D9EC0')
plt.grid(False)
plt.suptitle('Seasonal Reservations', y=1, fontsize=14, fontweight='bold')
plt.ylabel('Percentage (%)')
plt.show()
```

 __`New Feature 'Agent Category'`__

```python
#Function to determine the agent group, based on the total count
def agent_filtering(x, threshold):
    agent = x[0]
    count = x[1]
    
    if count >= threshold:
        return agent
    else:
        return 'Other'
```

```python
#Remove spaces in agent ID
df['Agent'] = df['Agent'].str.lstrip()

#Create a dataframe with the different agent ids and respective counts
df_agents = df['Agent'].value_counts().to_frame().reset_index()
df_agents.rename(columns = {'index':'agent_ID','Agent':'count'}, inplace = True)

#Define threshold for minimum value counts to keep agentID 
agent_threshold = initial_nrows * 0.05

#Get the categories
df_agents['Category'] = df_agents.apply(agent_filtering, args=[agent_threshold], axis=1)

#Reset index as the agent ID
df_agents.set_index('agent_ID',inplace=True)
df_agents.sample(n=8)
```

```python
#Dictionary with agent ID and respective category
agent_category = df_agents['Category'].to_dict()

#Get the agent IDs
agents = agent_category.keys()


#Replace the agent ID in the df with the category
for agent in agents:
    df.loc[df['Agent'] == agent,'AgentCategory'] = agent_category[agent]
    
df = df.drop(columns=['Agent'])    
```

 __`New Feature 'HasCompany'`__

```python
df.Company = df.Company.str.replace(' ', '')

# Fuction to help the creation of new Feature 'HasCompany'
def has_or_not_Company(row):
    if row['Company'] == 'NULL': # has no Company    
        return 0    
    else :
        return 1 # has Company 

df['HasCompany'] = df.apply(lambda row: has_or_not_Company(row), axis=1)

df = df.drop(columns=['Company'])
```

 __`New Feature 'ChangeOfRoom'`__

```python
df.ReservedRoomType = df.ReservedRoomType.str.replace(' ', '')
df.AssignedRoomType = df.AssignedRoomType.str.replace(' ', '')

# Fuction to help the creation of new Feature 'RoomTypeChange': to know if a RoomType was changed during the booking
def get_change_of_room(row):
    if row['ReservedRoomType'] == 'A' and row['AssignedRoomType'] == 'A' :     
        return 0 
    if row['ReservedRoomType'] == 'B' and row['AssignedRoomType'] == 'B' :     
        return 0
    if row['ReservedRoomType'] == 'C' and row['AssignedRoomType'] == 'C' :     
        return 0
    if row['ReservedRoomType'] == 'D' and row['AssignedRoomType'] == 'D' :     
        return 0
    if row['ReservedRoomType'] == 'E' and row['AssignedRoomType'] == 'E' :     
        return 0
    if row['ReservedRoomType'] == 'F' and row['AssignedRoomType'] == 'F' :     
        return 0
    if row['ReservedRoomType'] == 'G' and row['AssignedRoomType'] == 'G' :     
        return 0
    if row['ReservedRoomType'] == 'P' and row['AssignedRoomType'] == 'P' :     
        return 0
    else :
        return 1

df['RoomTypeChange'] = df.apply(lambda row: get_change_of_room(row), axis=1)

df = df.drop(columns=['AssignedRoomType'])
```

 __`New Feature 'PRTNationality'`__

```python
df.Country = df.Country.str.replace(' ', '')

# Fuction to help the creation of new Feature 'PRTNationality': 
def has_PRT_Nationality(row):
    if row['Country'] == 'PRT': # Have Portuguese Nationality     
        return 1    
    else :
        return 0 # Doesn't have Portuguese Nationality  

df['PRTNationality'] = df.apply(lambda row: has_PRT_Nationality(row), axis=1)

df = df.drop(columns=['Country'])
```

```python
#Potential revenue = ADR x Nr. StayInNights (StaysInWeekendNights + StaysInWeekNights)
df_solution_analysis['PotentialRevenue'] = (df_solution_analysis['StaysInWeekendNights']+ df_solution_analysis['StaysInWeekNights'])*df_solution_analysis['ADR']

#Actual Revenue
df_solution_analysis['ActualRevenue'] = -(df_solution_analysis['IsCanceled']-1)*df_solution_analysis['PotentialRevenue']

#Predict the Actual Revenue Based on the model's cancellation predictions
df_solution_analysis['PredictedRevenue'] = -(df_solution_analysis['IsCanceled_Prediction']-1)*df_solution_analysis['PotentialRevenue']

df_solution_analysis
```

```python
def cancelling_group(x):
    if x > 0.75:
        return 'Overbooking'
    elif x > 0.5 and x <=0.75:
        return 'Offer Perks'
    else:
        return 'Safe Bookings'
    
df_solution_analysis['Cancel_Class']=df_solution_analysis['Cancelling_Probability'].apply(cancelling_group)
df_solution_analysis
```
